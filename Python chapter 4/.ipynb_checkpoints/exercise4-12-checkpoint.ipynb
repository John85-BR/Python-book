```python
import random as rd

def move_tortoise():
    number_rd  = rd.randrange(1,11)
    
    if number_rd>=1 and number_rd<=5:
        return 3 #fast plod
    if number_rd==6 or number_rd==7:
        return -6 #slip
    if number_rd>=8 and number_rd<=10:
        return 1 #slow plod
    
def move_hare():
    number_rd  = rd.randrange(1,11)
    
    if number_rd==1 or number_rd==2:
        return 0 #sleep
    if number_rd==3 or number_rd==4:
        return 9 #big hop
    if number_rd==5:
        return -12 #big slip
    if number_rd>=6 and number_rd<=8:
        return 1 #small hop
    if number_rd==9 or number_rd==10:
        return -2 #small slip

def print_square(squares, pos_hare, pos_tortoise):
    

    if pos_hare>70:
        pos_hare=70
    
    if pos_tortoise>70:
        pos_tortoise=70
    
    if pos_hare==pos_tortoise:
        squares[pos_hare]='OUCH'
    else:
        squares[pos_hare]='H'
        squares[pos_tortoise]='T'
        
    return squares
    
squares = []

for i in range(71):
    squares+=' '
    
hare_pos = 1
tortoise_pos = 1
    
print('BANG!!!!\nAND THEY RE OFF!!!!!')

while True:
    
    movehare = move_hare()
    movetortoise = move_tortoise()
    
    if hare_pos+movehare>1 and hare_pos+movehare<=70:
        hare_pos += movehare
           
    if tortoise_pos+movetortoise>1 and tortoise_pos+movetortoise<=70:
        tortoise_pos += movetortoise
        
    if hare_pos>=70:
        print('\n\nhare wins')
        squares = print_square(squares, hare_pos,tortoise_pos) 
        break
    elif tortoise_pos>=70:
        print('\n\ntortoise wins')
        squares = print_square(squares, hare_pos,tortoise_pos)    
        break
    elif hare_pos>=70 and tortoise_pos>=70:
        print('It s a tie')
        squares = print_square(squares, hare_pos,tortoise_pos)
        break
        
    
    squares = print_square(squares, hare_pos,tortoise_pos)

for i in squares:
    print(i, end='')
```
