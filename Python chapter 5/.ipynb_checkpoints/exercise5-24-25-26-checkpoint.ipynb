```python
def initialize_deck():
    import random as rd
    
    num = ('Zero','Ace', 'Deuce','Three','Four','Five','Six',
              'Seven','Eight','Nine','Ten','Jack','Queen','King')
    deck = []
    cont = 0

    suit = ('Hearts','Diamonds','Clubs','Spades')
    suit_cont = 0

    while cont<4:
        for i in range(1,14):
            deck.append((num[i],suit[suit_cont]))

            if suit_cont<3:
                suit_cont+=1
            else:
                suit_cont=0

        cont+=1
    rd.shuffle(deck)
    return deck

import numpy as np

def print_deck(deck):
    cont=1
    for face,suit in deck:

        print(f'{face:<5} of {suit:20}',end='')
        if cont%4==0:
            print()
        cont+=1

def is_pair(hand):
    face = [i[0] for i in hand]
    values, frequencies = np.unique(face,return_counts = True)
    for i in frequencies:     
        if i >=2:
            return True
    return False

def is_two_pair(hand):
    face = [i[0] for i in hand]
    values, frequencies = np.unique(face,return_counts = True)
    if [1,2,2] == sorted(frequencies):
        return True
    else:
        return False
                
def is_three_of_a_kind(hand):
    face = [i[0] for i in hand]
    values, frequencies = np.unique(face,return_counts = True)
    for i in frequencies:     
        if i >=3:
            return True
    else:
        return False
                
def a_straight(hand):
    face = [i[0] for i in hand]
    num = ('Zero','Ace', 'Deuce','Three','Four','Five','Six',
              'Seven','Eight','Nine','Ten','Jack','Queen','King')
    cont = [num.index(i) for i in face]          
    cont_seq = 0
    for i in range(len(cont)-1):
        if cont[i]==(cont[i+1]-1):
            cont_seq+=1
            
    if cont_seq==4 or [1,10,11,12,13]==sorted(cont):
        return True
    else:
        return False
    
def a_flush(hand):
    suit = [i[1] for i in hand]
    values, frequencies = np.unique(suit,return_counts = True)
    for i in frequencies:     
        if i ==5:
            return True
    else:
        return False
    
    
def a_full_house(hand):
    suit = [i[1] for i in hand]
    values, frequencies = np.unique(suit,return_counts = True)
    if len(values)==2:
        return True
    else:
        return False
    
def four_of_a_kind(hand):
    face = [i[0] for i in hand]
    values, frequencies = np.unique(face,return_counts = True)
    if [1,4] == sorted(frequencies):
        return True
    else:
        return False
    
def straight_flush(hand):
    
    if not a_straight(hand):
        return False
    else:
        suit = [i[1] for i in hand]
        values, frequencies = np.unique(suit,return_counts = True)
        if len(values)==1:
            return True
        else:
            return False
      
    
def royal_flush(hand):
    if a_straight(hand) and a_flush(hand):
        face = [i[0] for i in hand]
        num = ('Zero','Ace', 'Deuce','Three','Four','Five','Six',
              'Seven','Eight','Nine','Ten','Jack','Queen','King')
        cont = [num.index(i) for i in face]  
        if [1,10,11,12,13]==sorted(cont):
            return True
    return False

def high_card(hand):
    face = [i[0] for i in hand]
    num = ('Zero','Ace', 'Deuce','Three','Four','Five','Six',
              'Seven','Eight','Nine','Ten','Jack','Queen','King')
    cont = [num.index(i) for i in face]
    return (num[max(cont)],max(cont))
        


deck = initialize_deck()

hands = [deck[i] for i in range(len(deck)) if i<5]

hands_test = [('Jack', 'Spades'), ('Deuce', 'Spades'), ('Three', 'Spades'), ('Four', 'Spades'), ('Five', 'Spades')]
print(hands_test)

print(f'Pair = {is_pair(hands_test)}')

print(f'Two Pair = {is_two_pair(hands_test)}')

print(f'Three of a kind = {is_three_of_a_kind(hands_test)}')

print(f'A straight = {a_straight(hands_test)}')

print(f'A flush = {a_flush(hands_test)}')

print(f'A full house = {a_full_house(hands_test)}')

print(f'four_of_a_kind = {four_of_a_kind(hands_test)}')

print(f'straight_flush = {straight_flush(hands_test)}')

print(f'royal_flush = {royal_flush(hands_test)}')

print(f'high_card = {high_card(hands_test)}')
```
