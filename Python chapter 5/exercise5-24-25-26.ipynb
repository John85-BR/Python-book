```python
def initialize_deck():
    import random as rd
    
    num = ('Zero','Ace', 'Deuce','Three','Four','Five','Six',
              'Seven','Eight','Nine','Ten','Jack','Queen','King')
    deck = []
    cont = 0

    suit = ('Hearts','Diamonds','Clubs','Spades')
    suit_cont = 0

    while cont<4:
        for i in range(1,14):
            deck.append((num[i],suit[suit_cont]))

            if suit_cont<3:
                suit_cont+=1
            else:
                suit_cont=0

        cont+=1
    rd.shuffle(deck)
    return deck

import numpy as np

def print_deck(deck):
    cont=1
    for face,suit in deck:

        print(f'{face:<5} of {suit:20}',end='')
        if cont%4==0:
            print()
        cont+=1

def is_pair(hand):
    face = [i[0] for i in hand]
    values, frequencies = np.unique(face,return_counts = True)
    for i in frequencies:     
        if i >=2:
            return True
    return False

def is_two_pair(hand):
    face = [i[0] for i in hand]
    values, frequencies = np.unique(face,return_counts = True)
    if [1,2,2] == sorted(frequencies):
        return True
    else:
        return False
                
def is_three_of_a_kind(hand):
    face = [i[0] for i in hand]
    values, frequencies = np.unique(face,return_counts = True)
    for i in frequencies:     
        if i >=3:
            return True
    else:
        return False
                
def a_straight(hand):
    face = [i[0] for i in hand]
    num = ('Zero','Ace', 'Deuce','Three','Four','Five','Six',
              'Seven','Eight','Nine','Ten','Jack','Queen','King')
    cont = [num.index(i) for i in face]          
    cont_seq = 0
    for i in range(len(cont)-1):
        if cont[i]==(cont[i+1]-1):
            cont_seq+=1
            
    if cont_seq==4 or [1,10,11,12,13]==sorted(cont):
        return True
    else:
        return False
    
def a_flush(hand):
    suit = [i[1] for i in hand]
    values, frequencies = np.unique(suit,return_counts = True)
    for i in frequencies:     
        if i ==5:
            return True
    else:
        return False
    
    
def a_full_house(hand):
    suit = [i[1] for i in hand]
    values, frequencies = np.unique(suit,return_counts = True)
    if [2,3] == sorted(frequencies):
        return True
    else:
        return False
    
def four_of_a_kind(hand):
    face = [i[0] for i in hand]
    values, frequencies = np.unique(face,return_counts = True)
    if [1,4] == sorted(frequencies):
        return True
    else:
        return False
    
def straight_flush(hand):
    
    if not a_straight(hand):
        return False
    else:
        suit = [i[1] for i in hand]
        values, frequencies = np.unique(suit,return_counts = True)
        if len(values)==1:
            return True
        else:
            return False
      
    
def royal_flush(hand):
    if a_straight(hand) and a_flush(hand):
        face = [i[0] for i in hand]
        num = ('Zero','Ace', 'Deuce','Three','Four','Five','Six',
              'Seven','Eight','Nine','Ten','Jack','Queen','King')
        cont = [num.index(i) for i in face]  
        if [1,10,11,12,13]==sorted(cont):
            return True
    return False

def high_card(hand):
    face = [i[0] for i in hand]
    num = ('Zero','Ace', 'Deuce','Three','Four','Five','Six',
              'Seven','Eight','Nine','Ten','Jack','Queen','King')
    cont = [num.index(i) for i in face]
    return (max(cont),num[max(cont)])
        

def verify_game_hand(hand):
        
    if royal_flush(hand):
        return (10,'Royal Flush')
    elif straight_flush(hand):
        return (9,'Straight Flush')
    elif four_of_a_kind(hand):
        return (8,'Four of a kind')
    elif a_full_house(hand):
        return (7, 'Full House')
    elif a_flush(hand):
        return (6, 'Flush')
    elif a_straight(hand):
        return (5, 'A Straight')
    elif is_three_of_a_kind(hand):
        return (4,'Three of a kind')
    elif is_two_pair(hand):
        return (3,'Two pair')
    elif is_pair(hand):
        return (2, 'A pair')
    else:
        return (1, 'High Card')
    
    
deck = initialize_deck()

hands1 = [deck.pop() for i in range(3)]

hands2 = [deck.pop() for i in range(3)]   

while True:
        
    print(f'Hands one: {hands1}') 
    points_hands1, game_hands1 = verify_game_hand(hands1)
    print(f'{game_hands1}')
    print(f'Hands two: {hands2}')
    points_hands2, game_hands2 = verify_game_hand(hands2)
    print(f'{game_hands2}')

    input('Tecle algo para continuar')
    
    hands1 += [deck.pop() for i in range(1)]

    hands2 += [deck.pop() for i in range(1)]  
    
    if len(hands1)==6:
        if points_hands1>points_hands2:
            print(f' Hands 1 wins with a {game_hands1}')
        elif points_hands1<points_hands2:
            print(f'Hands 2 wins with a {game_hands2}')
        elif points_hands1==points_hands2:
            if points_hands1== 1 and points_hands2==1:
                point1, high1 = high_card(hands1)
                point2, high2 = high_card(hands2)
                if point1>point2:
                    print(f'Hands 1 wins with a high card: {high1} beats {high2}')
                elif point1<point2:
                    print(f'Hands 2 wins with a high card: {high2} beats {high1}')
                else:
                    print(f'draw {high1} x {high2}')
            else:
                print(f'draw {game_hands1} x {game_hands2}')         
        break







```
