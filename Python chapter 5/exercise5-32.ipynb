```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import random
import seaborn as sns
```

```python
die1 = [random.randrange(1, 7) for i in range(360)]
die2 = [random.randrange(1, 7) for i in range(360)]
die_sum = [(x+y) for x,y in zip(die1,die2)]
values, frequencies = np.unique(die_sum, return_counts=True)
```

```python
axes = sns.barplot(y=values, x=frequencies, orient='h')
axes.set(ylabel='Die Value', xlabel='Frequency') 
title = f'Rolling a coin {len(die_sum):,} Times'
axes.set_title(title)

for bar, frequency in zip(axes.patches, frequencies):
    text_x = bar.get_width()
    text_y = bar.get_y()+bar.get_height()
    text = f'{frequency}\n{frequency / len(die_sum):.1%}'
    axes.text(text_x, text_y, text, fontsize=11)
```

```python
%recall 2
```
