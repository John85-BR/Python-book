```python
# fig04_02.py
"""Simulating the dice game Craps."""
import random

def roll_dice():
    """Roll two dice and return their face values as a tuple."""
    die1 = random.randrange(1, 7)
    die2 = random.randrange(1, 7)
    return (die1, die2)  # pack die face values into a tuple

def display_dice(dice):
    """Display one roll of the two dice."""
    die1, die2 = dice  # unpack the tuple into variables die1 and die2
    print(f'Player rolled {die1} + {die2} = {sum(dice)}')

games = 0

dict_won = {}
dict_lost = {}
while games<100000:
    
    rolls = 0
    
    die_values = roll_dice()  # first roll
    #display_dice(die_values)
    rolls+=1
    # determine game status and point, based on first roll
    sum_of_dice = sum(die_values)

    if sum_of_dice in (7, 11):  # win
        game_status = 'WON'
        if rolls in dict_won:
            dict_won[rolls] += 1
        else:
            dict_won[rolls] = 1
    elif sum_of_dice in (2, 3, 12):  # lose
        game_status = 'LOST'
        if rolls in dict_lost:
            dict_lost[rolls] += 1
        else:
            dict_lost[rolls] = 1
    else:  # remember point
        game_status = 'CONTINUE'
        my_point = sum_of_dice
        #print('Point is', my_point)

    # continue rolling until player wins or loses
    while game_status == 'CONTINUE':
        die_values = roll_dice()
        #display_dice(die_values)
        sum_of_dice = sum(die_values)
        rolls+=1

        if sum_of_dice == my_point:  # win by making point
            game_status = 'WON'
            if rolls in dict_won:
                dict_won[rolls] += 1
            else:
                dict_won[rolls] = 1
        elif sum_of_dice == 7:  # lose by rolling 7
            game_status = 'LOST'
            if rolls in dict_lost:
                dict_lost[rolls] += 1
            else:
                dict_lost[rolls] = 1
            
   # if game_status == 'WON':
        #print('Player wins')
    #else:
        #print('Player loses')
            
    games+=1
    
sum_of_won = 0
sum_of_lost = 0
# display "wins" or "loses" message
for k, value in dict_won.items():
    sum_of_won+=value
    
for k, value in dict_lost.items():
    sum_of_lost+=value
    
percentage_wins = (sum_of_won/games)*100
percentage_losses = (sum_of_lost/games)*100
print(f'Percentage of wins: {percentage_wins:.2f}%')
print(f'Percentage of losses: {percentage_losses:.2f}%')


print('Percentage os wins/losses based on total number of rolls')
cumulate = 0
print(f'\t% Resolved\t\tCumulate %')
print(f'Rolls\ton this roll\t of games revolved')

max_win_or_lost = max(dict_won.keys()) if max(dict_won.keys())>= max(dict_lost.keys()) else max(dict_lost.keys())
for roll in range(1,max_win_or_lost+1):
    if roll in dict_won and roll in dict_lost:
        percentage_wins_roll = ((dict_won[roll]/games)*100)+((dict_lost[roll]/games)*100)
        cumulate+=percentage_wins_roll
    elif roll in dict_won and roll not in dict_lost:
        percentage_wins_roll = (dict_won[roll]/games)*100
    elif roll not in dict_won and roll in dict_lost:
        percentage_wins_roll = (dict_lost[roll]/games)*100
    print(f'{roll:>5}\t{percentage_wins_roll:>12.2f}\t{cumulate:>18.2f}')
    
```
