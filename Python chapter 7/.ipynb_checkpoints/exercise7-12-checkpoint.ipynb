```python
import numpy as np
import random as rd

def check_winner_horizontal(array, char):
    for i in range(1,5):
        if np.all(array[(i-1):i]==char):
            return True
        if np.all(array[: , (i-1):i]==char):
            return True
    
    if array[0][0]==char and array[1][1]==char and array[2][2]==char and array[3][3]==char:
            return True
    if array[0][3]==char and array[1][2]==char and array[2][1]==char and array[3][0]==char:
            return True
               
    return False

def check_winner_array3d(array_3d, char):
    
    for i in range(4):
        if check_winner_horizontal(array_3d[i],char):
            return True
    #vetical entre tabelas
    for row in range(4):
        for col in range(4):
            if array_3d[0][row][col] == char and array_3d[1][row][col] == char and array_3d[2][row][col] == char and array_3d[3][row][col] == char:
                return True
    #diagonal por linha entre tabelas        
    for row in range(4):
        for col in range(4):
            if array_3d[0][row][0] == char and array_3d[1][row][1] == char and array_3d[2][row][2] == char and array_3d[3][row][3] == char:
                return True
    #digonal entre colunas e linhas por tabela
    if array_3d[0][0][0] == char and array_3d[1][1][1] == char and array_3d[2][2][2] == char and array_3d[3][3][3] == char:
                return True
        
    if array_3d[0][0][3] == char and array_3d[1][1][2] == char and array_3d[2][2][1] == char and array_3d[3][3][0] == char:
                return True
    
    return False        

array_3d = np.full((4,4,4),' ')
char = 'X'

cont = 0
table_full = False
has_winner = False
   
while not np.all(table_full) and not has_winner:
    
    print(array_3d)
    print('-------------------------------------')
    
    while True and not has_winner:#Robô 1
        player1 = (rd.randrange(0,4),rd.randrange(0,4),rd.randrange(0,4))
        if array_3d[player1] == ' ':        
            array_3d[player1] = 'X'
            if check_winner_array3d(array_3d,'X'):
                print('Player1 winner')
                has_winner = True
            break
    
    while True and not has_winner:#Robô 2
        player2 = (rd.randrange(0,4),rd.randrange(0,4),rd.randrange(0,4))
        if array_3d[player2] == ' ':        
            array_3d[player2] = 'O'  
            if check_winner_array3d(array_3d,'O'):
                print('Player2 winner')
                has_winner = True
            break
    table_full = array_3d!=' ' 

if np.all(table_full) and not has_winner:
    print('draw')
    
print(array_3d)

```
