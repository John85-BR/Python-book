```python
import numpy as np
import random as rd

chess_board = np.zeros((8,8),dtype=int)
heuristic_board = np.array([[2,3,4,4,4,4,3,2],
                            [3,4,6,6,6,6,4,3],
                            [4,6,8,8,8,8,6,4],
                            [4,6,8,8,8,8,6,4],
                            [4,6,8,8,8,8,6,4],
                            [4,6,8,8,8,8,6,4],
                            [3,4,6,6,6,6,4,3],
                            [2,3,4,4,4,4,3,2]])
horizontal = np.array([2,1,-1,-2,-2,-1,1,2])
vertical = np.array([-1,-2,-2,-1,1,2,2,1])
dict_heuristic = {}

for i in range(0,8):
    for j in range(0,8):
        
        chess_board = np.zeros((8,8),dtype=int)
        current_row = i
        current_column = j
        counter = 1
        heuristic = 0
        point = 1
        chess_board[current_row][current_column] = point
        

        while counter<65:

            counter+=1
            for move_number in range(0,8):    

                current_row += vertical[move_number]
                current_column += horizontal[move_number]

                if (current_row>=0 and current_row<8) and (current_column>=0 and current_column<8):
                    if chess_board[current_row][current_column]==0:
                        heuristic = heuristic_board[current_row][current_column]
                        dict_heuristic[heuristic] = [current_row,current_column]
                        key_min = min(dict_heuristic.keys())
                        current_row -= vertical[move_number]
                        current_column -= horizontal[move_number]
                    else:
                        current_row -= vertical[move_number]
                        current_column -= horizontal[move_number]
                else:            
                    current_row -= vertical[move_number]
                    current_column -= horizontal[move_number]

            if key_min in dict_heuristic:
                v, h = dict_heuristic[key_min]
                current_row = v
                current_column = h
                point+=1
                chess_board[current_row][current_column] = point 
                dict_heuristic.clear()

        print(chess_board, point)
        if point==64:
            print(f'Completed 64 moves when the horse starts at ({i}:{j}) location')
        print('-------------------------')

    
    

```
