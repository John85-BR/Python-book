```python
import numpy as np
import random as rd  
from collections import Counter
import pandas as pd

horizontal = np.array([2,1,-1,-2,-2,-1,1,2])
vertical = np.array([-1,-2,-2,-1,1,2,2,1])        
lenght = np.zeros(1000000,dtype=int)


for i in range(1000000):

    chess_board = np.zeros((8,8),dtype=int)
    current_row = 0
    current_column = 0
    counter = 1
    point = 1
    chess_board[current_row][current_column] = point
    
    while counter<65:

        counter+=1

        move_number = rd.randrange(0,8)
        current_row += vertical[move_number]
        current_column += horizontal[move_number]

        if (current_row>=0 and current_row<8) and (current_column>=0 and current_column<8):
            if chess_board[current_row][current_column]==0:
                point+=1
                chess_board[current_row][current_column] = point
            else:
                current_row -= vertical[move_number]
                current_column -= horizontal[move_number]
        else:            
            current_row -= vertical[move_number]
            current_column -= horizontal[move_number]

    #print(chess_board, point)
    lenght[i]+=point
    if i%100000==0:
        print(f'At {i}')
    #print('-------------------------')

    
counter = Counter(lenght)

grades = pd.Series(counter)
print(grades.sort_values(ascending = False))

```
